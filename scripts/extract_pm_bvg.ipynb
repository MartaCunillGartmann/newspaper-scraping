```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

# Starte den Browser
driver = uc.Chrome()
driver.get("https://www.bvg.de/de/unternehmen/medienportal/pressemitteilungen")
time.sleep(5)  # Warte, bis die Seite vollstÃ¤ndig geladen ist

all_links = set()

while True:
    # Sammle die Links auf der aktuellen Seite
    links = driver.find_elements(By.CSS_SELECTOR, "a[href^='/de/unternehmen/medienportal/pressemitteilungen/']")
    for link in links:
        href = link.get_attribute("href")
        if href:
            all_links.add(href)
    
    # Versuche, den "NÃ¤chste Seite"-Button zu finden und klicke ihn
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "a.pagination__link--next")
        next_button.click()
        time.sleep(5)  # Warte, bis die nÃ¤chste Seite geladen ist
    except NoSuchElementException:
        print("Keine weitere Seite gefunden.")
        break

# Gib die gesammelten Links aus
print(f"Insgesamt {len(all_links)} Links gefunden:")
for link in all_links:
    print(link)

driver.quit()
```

```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = uc.Chrome()
driver.get("https://www.bvg.de/de/unternehmen/medienportal/pressemitteilungen")
time.sleep(5)

all_links = set()
current_page = 1

while True:
    print(f"ğŸ“„ Seite {current_page}")
    time.sleep(3)

    # Links auf dieser Seite sammeln
    links = driver.find_elements(By.CSS_SELECTOR, "a[href^='/de/unternehmen/medienportal/pressemitteilungen/']")
    for link in links:
        href = link.get_attribute("href")
        if href:
            all_links.add(href)

    try:
        next_page_number = str(current_page + 1)

        # Warte, bis der Button sichtbar ist
        wait = WebDriverWait(driver, 10)
        next_button = wait.until(EC.presence_of_element_located(
            (By.XPATH, f"//button[normalize-space(text())='{next_page_number}']")))

        # Scroll zum Button
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", next_button)
        time.sleep(1)

        # Klicken mit ActionChains (fÃ¼r mehr ZuverlÃ¤ssigkeit)
        ActionChains(driver).move_to_element(next_button).click().perform()

        current_page += 1
        time.sleep(4)

    except Exception as e:
        print(f"âœ… Keine weitere Seite gefunden oder Fehler bei Seite {current_page}: {e}")
        break

# Ergebnisse anzeigen
print(f"\nğŸ”— Insgesamt {len(all_links)} Links gefunden:")
for link in all_links:
    print(link)

driver.quit()
```

```python
# Download der Pressemitteilungen als HTML-Dateien
import requests
import os

# Zielordner
SAVE_FOLDER = os.path.join("input", "pm_bvg_raw")
os.makedirs(SAVE_FOLDER, exist_ok=True)
```

```python
print("Ordnerpfad:", SAVE_FOLDER)
print("Existiert Ordner?", os.path.exists(SAVE_FOLDER))
print("Dateien im Ordner:", os.listdir(SAVE_FOLDER))
```

```python
# Lade jede Pressemitteilung herunter
for url in all_links:
    try:
        # Dateiname aus dem Link extrahieren
        slug = url.strip("/").split("/")[-1]
        filename = f"{slug}.html"
        save_path = os.path.join(SAVE_FOLDER, filename)
          # Nur herunterladen, wenn Datei noch NICHT existiert
        if os.path.exists(save_path):
            print(f"Bereits vorhanden, Ã¼berspringe: {filename}")
            continue

        # Inhalt herunterladen und speichern
        response = requests.get(url)
        response.raise_for_status()
        with open(save_path, "w", encoding="utf-8") as f:
            f.write(response.text)

        print(f"Gespeichert: {filename}")
    except Exception as e:
        print(f"Fehler bei {url}: {e}")
```
